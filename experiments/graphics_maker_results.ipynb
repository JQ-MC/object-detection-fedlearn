```python
# Library imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path routes definition
path = "./exp_noniid"
client_fed = "/fed_noniid_c3.csv"
client_cent = "/cen_noniid_c3.csv"
```

```python
# Data reading
data_fed = pd.read_csv(path+client_fed)
data_cent = pd.read_csv(path+client_cent)
```

```python
# Data checking
data_fed.head()
data_fed.shape
data_cent.head()
data_cent.shape
```

```python
# creation of New variable for the x axis
epoch = [x for x in range(1,401)]
data_fed["epoch"] = epoch
data_cent["epoch"] = epoch
```

### Plots

```python
fig, axs = plt.subplots(5, 1)

#val/box_loss
axs[0].plot(data_fed["epoch"], data_fed['        val/box_loss'], data_cent["epoch"], data_cent['        val/box_loss'])
axs[0].set_xlabel('epochs', fontsize=12)
axs[0].set_ylabel('val/box_loss', fontsize=12)
axs[0].set_title("Error d'ajust de bounding-box en validació (box_loss)", fontsize=16)
axs[0].grid(True)
axs[0].legend(['Federat', 'Centralitzat'],prop={"size":14})
axs[0].annotate('Pèrdua de connexió', xy=(265, 0.042),  
            xytext=(0.82, 0.7), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )
axs[0].annotate('Reestabliment de connexió', xy=(320, 0.04),  
            xytext=(0.88, 0.4), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )

# val/cls_loss
axs[1].plot(data_fed["epoch"], data_fed['        val/cls_loss'], data_cent["epoch"], data_cent['        val/cls_loss'])
axs[1].set_xlabel('epochs', fontsize=12)
axs[1].set_ylabel('val/cls_loss', fontsize=12)
axs[1].set_title("Error de predicció de classe en validació (cls_loss)", fontsize=16)
axs[1].grid(True)
axs[1].legend(['Federat', 'Centralitzat'],prop={"size":14})
axs[1].annotate('Pèrdua de connexió', xy=(265, 0.052),  
            xytext=(0.82, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )
axs[1].annotate('Reestabliment de connexió', xy=(320, 0.04),  
            xytext=(0.68, 0.2), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )

# val/obj_loss
axs[2].plot(data_fed["epoch"], data_fed['        val/obj_loss'], data_cent["epoch"], data_cent['        val/obj_loss'])
axs[2].set_xlabel('epochs', fontsize=12)
axs[2].set_ylabel('val/obj_loss', fontsize=12)
axs[2].set_title("Error de predicció de nombre d'objectes en validació (obj_loss)", fontsize=16)
axs[2].grid(True)
axs[2].legend(['Federat', 'Centralitzat'],prop={"size":14})
axs[2].annotate('Pèrdua de connexió', xy=(265, 0.034),  
            xytext=(0.82, 0.7), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )
axs[2].annotate('Reestabliment de connexió', xy=(320, 0.027),  
            xytext=(0.75, 0.1), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )

# '     metrics/mAP_0.5'
axs[3].plot(data_fed["epoch"], data_fed['     metrics/mAP_0.5'], data_cent["epoch"], data_cent['     metrics/mAP_0.5'])
axs[3].set_xlabel('epochs', fontsize=12)
axs[3].set_ylabel('metrics/mAP_0.5', fontsize=12)
axs[3].set_title("Mean Average Precision 0.5 (mAP 0.5) ", fontsize=16)
axs[3].grid(True)
axs[3].legend(['Federat', 'Centralitzat'],prop={"size":14})
axs[3].annotate('Pèrdua de connexió', xy=(265, 0.34),  
            xytext=(0.84, 0.14), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )
axs[3].annotate('Reestabliment de connexió', xy=(320, 0.6),  
            xytext=(0.7, 0.8), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )


# 'metrics/mAP_0.5:0.95'
axs[4].plot(data_fed["epoch"], data_fed['metrics/mAP_0.5:0.95'], data_cent["epoch"], data_cent['metrics/mAP_0.5:0.95'])
axs[4].set_xlabel('epochs', fontsize=12)
axs[4].set_ylabel('metrics/mAP_0.5:0.95', fontsize=12)
axs[4].set_title("Mean Average Precision 0.5:0.95 (mAP 0.5:0.95) ", fontsize=16)
axs[4].grid(True)
axs[4].legend(['Federat', 'Centralitzat'],prop={"size":14})
axs[4].annotate('Pèrdua de connexió', xy=(265, 0.18),  
            xytext=(0.84, 0.14), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )
axs[4].annotate('Reestabliment de connexió', xy=(320, 0.4),  
            xytext=(0.92, 0.4), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )

plt.subplots_adjust(
                    bottom=0.1, 
                    top=0.93, 
                    hspace=0.6)
fig.set_size_inches(12.5, 16.5)
fig.suptitle("Mètriques d'entrenament del client 3 amb dades no-iid", fontsize=20)
plt.show()
```
